# SEASON 2021-2022 : Current season

```python
Form_21_22 = "LLLWWWDDWWWLWLLWWWWLDWW"
print(Form_21_22)
```

```python
List_21_22 = list(Form_21_22)
print(List_21_22)
```

```python
def convert(Lis) :
    list1 = []
    for i in range(len(Lis)):
        if Lis[i] == 'W':
             list1.append(3)
        elif Lis[i] == 'L':
             list1.append(0)
        else :
             list1.append(1)
    return list1
list1 = convert(List_21_22)
list1
```

```python
type(list1[1])
```

```python

def count(liste) :
    c = 0
    for i in range(len(liste)):
        sum = 0
        if i + 15 < len(liste):
            for j in range(i,i+15):
                sum = sum + liste[j]
            if ( sum >= 30):
                c = c +1
                print("|Matchday " + str(i+1) + " to Matchday " +str(j+1) + "| Points accumulated = " + str(sum))
    print(c)

count(list1)
```

As we can see that the run of games are just next to each other so we combine and calculate the avg points

```python
import numpy as np
a = np.array(list1)
result = np.sum(a[3:19])
print("Points accumulated from game no 4 to 19 = " + str(result))
print("Avg Points per game in this run = " + str(round(result/16,2)))
```

#### Final Plotting

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,3), dpi=200)
ax = plt.subplot()
plt.ylim(4,-1)
plt.xlim(0,26)

matches = [x for x in range(1,len(list1)+1)]




ax.yaxis.set_ticks([0,1,3])
ax.xaxis.set_ticks([x for x in range(0,len(list1)+3,3)])
ax.set_yticklabels(['Loss','Draw','Win'])


for y,z in zip(matches,list1): 
    if z == 0:
        plt.scatter(y,z,s = 40, alpha = 1,color= "red",zorder = 3)
    elif z == 3:
        plt.scatter(y,z,s = 40, alpha = 1,color= "green",zorder = 3)
    else:
        plt.scatter(y,z,s = 40, alpha = 1,color= "blue",zorder = 3)
            
            
            



#lines
xpoints = matches
ypoints = list1
plt.plot(xpoints, ypoints, color = 'black',linestyle = '--',zorder=2)


plt.xlabel('Match') #Note 2017 denotes the season of 2016-2017
plt.ylabel('Outcome')
plt.title("Arsenal Form 2021-22 ( till match 23)")

plt.gca().invert_yaxis()

# displaying the runs where the avg accumulated points > 2
xpoints = [4,19]
ypoints = [3.3,3.3]
plt.plot(xpoints, ypoints, color = 'black',linestyle = '-',zorder=2)

fig.text(
        x=0.35, y=.8, s='Match 4 to 19 | PPG = 2.19',
        ha='left',va='bottom',size=10,fontname="AppleGothic",fontweight="bold",
        color="black"
    )


# Show the plot
plt.show()
```
